# **Model Training**

# Train Test Split

```python
df
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['Species','target'],axis=1)
y = df['Species']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2 ,random_state = 41)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

# **Multi-Class Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
lr_model = LogisticRegression(solver = 'saga',multi_class='ovr',max_iter=100)
```

# **GridSearch for Best Hyper-Parameter**

Main parameters choices are regularization penalty choice and regularization **c** value

```python
penalty = ['l1','l2','elasticnet']
l1_ratio = np.linspace(0,1,20)

# use algorithimically spaced c value(recomended in official docs)
C = np.logspace(0,10,20)
#By using np.logspace, you ensure that the values are logarithmically evenly distributed. This means that there will be an equal "density" of values across different orders of magnitude.

parameters = {
    'penalty':penalty,
    'l1_ratio':l1_ratio,
    'C': C
}
```

```python
grid_model = GridSearchCV(lr_model,parameters)
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_params_
```

#**Model Performance on Classification Tasks**

```python
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score,f1_score,ConfusionMatrixDisplay,confusion_matrix
```

```python
y_pred = grid_model.predict(x_test)
y_pred
```

```python
accuracy_score = accuracy_score(y_test,y_pred)
accuracy_score
```

```python
precision_score = precision_score(y_test,y_pred,average = 'macro')
precision_score
```

```python
recall_score = recall_score(y_test,y_pred,average = 'macro')
recall_score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
display = ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = grid_model.classes_)
display.plot()
plt.show()
```

From the chart above, the model is predicting pretty good.

**There are some wrong points on versicolor where is predicted as setosa, versicolor where is predicted as virginica and virgina where is predicted as versicolor.**

```python
print(classification_report(y_test,y_pred))
```

# **Evaluating Curves and AUC**

**OvR: One to Rest**

```python
grid_model.predict_proba(x_test)
```

```python
from sklearn.metrics import RocCurveDisplay

fig = plt.figure()
ax = fig.add_subplot(111)

for i, class_id in enumerate(grid_model.classes_):
  y_test_binary = (y_test == class_id).astype(int)
  #  For each iteration, we create a new binary y_test. If an actual label in y_test matches the class_id of the current iteration, it's marked as 1 (positive); otherwise, it's 0 (negative).

  y_score_current_class = grid_model.predict_proba(x_test)[:,i]
  # We get the predicted probabilities for all classes for each sample in x_test. This is crucial because ROC curves need probabilities, not just hard class predictions.

  RocCurveDisplay.from_predictions(y_test_binary,y_score_current_class,ax=ax,)
  # switched to from_predictions for clarity because you're directly providing the y_true (binarized) and y_score (probabilities). While from_estimator can sometimes infer probabilities, from_predictions is more direct when you've already computed them.


plt.show()
```

 * we can see that 1 class(blue) is perfect on y_axis.

 * And another 2(orange and green) are not perfect because there are misclassifications.

 * But in reality we can't expect ROC curve will be perfect
